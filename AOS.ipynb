```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# List of CSV file names (make sure these match what you uploaded)
file_paths = [
    "output0.csv", "output1.csv", "output2.csv",
    "output3.csv", "output4.csv", "output5.csv",
    "output6.csv", "output7.csv", "output8.csv"
]

# Load and tag each DataFrame
dataframes = []
for idx, path in enumerate(file_paths):
    df = pd.read_csv(path)
    df["Simulation"] = f"Sim_{idx}"
    dataframes.append(df)

# Combine all into one DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Aggregate statistics
summary_stats = combined_df.groupby("Simulation").agg({
    "Avg Queueing Time": "mean",
    "CPU Load": "mean",
    "99.9% Tail Latency": "mean",
    "95% Tail Latency": "mean",
    "Task Load": "mean",
    "Work Steal Load": "mean"
}).reset_index()

summary_stats.columns = [
    "Simulation", "Avg_Queueing_Time", "Avg_CPU_Load",
    "Avg_Tail_Latency_99_9", "Avg_Tail_Latency_95",
    "Avg_Task_Load", "Avg_Work_Steal_Load"
]

# Plotting
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=summary_stats, x="Avg_CPU_Load", y="Avg_Queueing_Time", marker="o", hue="Simulation", palette="tab10")

plt.title("Latency vs CPU Utilization across Simulations")
plt.xlabel("Average CPU Load (%)")
plt.ylabel("Average Queueing Time (ns)")
plt.legend(title="Simulation", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File names must match exactly what you uploaded
file_paths = [
    "output0.csv", "output1.csv", "output2.csv",
    "output3.csv", "output4.csv", "output5.csv",
    "output6.csv", "output7.csv", "output8.csv"
]

# Read and tag each file
dataframes = []
for idx, path in enumerate(file_paths):
    df = pd.read_csv(path)
    df["Simulation"] = f"Sim_{idx}"
    dataframes.append(df)

# Combine into one DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Summary statistics
summary_stats = combined_df.groupby("Simulation").agg({
    "Avg Queueing Time": "mean",
    "CPU Load": "mean",
    "99.9% Tail Latency": "mean",
    "95% Tail Latency": "mean",
    "Task Load": "mean",
    "Work Steal Load": "mean"
}).reset_index()

summary_stats.columns = [
    "Simulation", "Avg_Queueing_Time", "Avg_CPU_Load",
    "Avg_Tail_Latency_99_9", "Avg_Tail_Latency_95",
    "Avg_Task_Load", "Avg_Work_Steal_Load"
]
```

Plot: Latency vs CPU Utilization


```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=summary_stats, x="Avg_CPU_Load", y="Avg_Queueing_Time", marker="o", hue="Simulation", palette="tab10")
plt.title("Latency vs CPU Utilization")
plt.xlabel("Average CPU Load (%)")
plt.ylabel("Average Queueing Time (ns)")
plt.legend(title="Simulation", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```
